```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
sales = pd.read_csv("home_data.csv")
```

```python
sales.head()
```

# Regression model with single feature

```python
X = sales['sqft_living'].values
y = sales['price'].values
#converting 1-D to 2-D
X = np.array(X).reshape(-1,1)
y = np.array(y).reshape(-1,1)
```

```python
colors = np.random.rand(21613)       #generate random color codes
plt.scatter(X, y, c=colors, alpha=0.5)
plt.title('Scatter plot')
plt.xlabel('Square foot living')
plt.ylabel('Price')
plt.show()
```

```python
#splitting the data into training and test 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
#linear regression object
sqft_model = LinearRegression()
#training the model
sqft_model.fit(X_train, y_train)
```

```python
#coefficients of the model
print('Coefficients: \n', sqft_model.coef_)
#intercept of the model
print('Intercept: \n', sqft_model.intercept_)
```

```python
price_pred = sqft_model.predict(X_test)
```

```python
#predicted price -- square foot value -- real price
for x in range(len(price_pred)):
  print(price_pred[x], X_test[x], y_test[x])
```

```python
#r-squared value
r2_score(y_test, price_pred)
```

```python
print('Mean Absolute Error:', mean_absolute_error(y_test, price_pred))  
print('Mean Squared Error:', mean_squared_error(y_test, price_pred))  
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, price_pred)))
```

```python
plt.plot(X_test, y_test, '.', X_test, price_pred, 'r--')
```

# Regression model with multiple features

```python
my_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode']
```

```python
sales[my_features].head()
```

```python
zipcode = sales['zipcode']
zipcode = np.array(zipcode).reshape(-1,1)
plt.figure(figsize=(200,10))
plt.xticks(np.arange(min(zipcode), max(zipcode)+1))  #setting the range for 200 zipcodes
plt.scatter(zipcode, y, c=colors, alpha=0.3)
plt.title('Scatter plot')
plt.xlabel('Zip code')
plt.ylabel('Price')
plt.show()
```

```python
X = sales[my_features].values
y = sales['price'].values
X = np.array(X).reshape(-1,6)
y = np.array(y).reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
my_features_model = LinearRegression()
my_features_model.fit(X_train, y_train)
```

```python
price_pred = my_features_model.predict(X_test)
```

```python
#predicted price -- actual price
for x in range(len(price_pred)):
    print(price_pred[x], '--',  y_test[x])
```

```python
r2_score(y_test, price_pred)
```

```python
print('Mean Absolute Error:', mean_absolute_error(y_test, price_pred))  
print('Mean Squared Error:', mean_squared_error(y_test, price_pred))  
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, price_pred)))
```

# Advanced features for a better model

```python
advanced_features = [
'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode',
'condition', # condition of house
'grade', # measure of quality of construction
'waterfront', # waterfront property
'view', # type of view
'sqft_above', # square feet above ground
'sqft_basement', # square feet in basement
'yr_built', # the year built
'yr_renovated', # the year renovated
'lat', 'long', # the lat-long of the parcel
'sqft_living15', # average sq.ft. of 15 nearest neighbors 
'sqft_lot15', # average lot size of 15 nearest neighbors 
]
```

```python
X = sales[advanced_features].values
y = sales['price'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
advanced_model = LinearRegression()
advanced_model.fit(X_train, y_train)
```

```python
price_pred = advanced_model.predict(X_test)
```

```python
print('Mean Absolute Error:', mean_absolute_error(y_test, price_pred))  
print('Mean Squared Error:', mean_squared_error(y_test, price_pred))  
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, price_pred)))
```

```python
r2_score(y_test, price_pred)
```
